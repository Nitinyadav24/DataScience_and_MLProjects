```
from nltk.corpus import inaugural
```

```

nltk.download('inaugural')
```

```
inaugural.fileids()
```

```
for speech in inaugural.fileids():
    word_tot=len(inaugural.words(speech))
    print (word_tot,speech)
```

```
speech_len=[(len(inaugural.words(speech)),speech) for speech in inaugural.fileids()]
```

```
max(speech_len)
```

```
speech_len
```

```
min(speech_len)
```

```
for speech in inaugural.fileids():
    word_len=len(inaugural.words(speech))
    sents_len=len(inaugural.sents(speech))
    print(word_len, sents_len)
```
