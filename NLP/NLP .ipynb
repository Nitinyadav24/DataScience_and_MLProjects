```
import nltk
```

```
from nltk.corpus import inaugural
```

```
nltk.download('inaugural')
nltk.download('punkt')
```

```
inaugural.fileids()
```

```
for speech in inaugural.fileids():
  Total_word=len(inaugural.words(speech))
  print(Total_word,speech)
```

```
# Find the longest speech 
Speech_len=[(len(inaugural.words(speech)),speech) for speech in inaugural.fileids()]
Speech_len
```

```
max(Speech_len)
```

```
# Find the average word per speech 
for speech in inaugural.fileids():
  Words=len(inaugural.words(speech))
  Sen_tot=len(inaugural.sents(speech))
  print(round(Words/Sen_tot,0),speech)
```

```
# plotting the word per speech against year
import pandas as pd
```

```
df=pd.DataFrame([int(speech[:4]),len(inaugural.words(speech))/len(inaugural.sents(speech))] for speech in inaugural.fileids())
```

```
df.columns=['Year','WPS']
```
