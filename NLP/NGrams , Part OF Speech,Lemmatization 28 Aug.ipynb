```
from nltk.util import ngrams
```

```
text='if nce i will go to the beqach'
token=nltk.word_tokenize(text)
bigram=ngrams(token, 2)
for i in bigram:
  print(i)
```

```
# for 4 items
fourgram=ngrams(token,4)
for j in fourgram:
  print(j)
```

```
def n_gram(text,n):
  token=nltk.word_tokenize(text)
  grams=ngrams(token,n)
  return grams
```
