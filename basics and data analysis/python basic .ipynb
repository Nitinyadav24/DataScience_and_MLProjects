# Python Sorting

```
strs = ['A','D','B','C']
print(sorted(strs))  
print(sorted(strs, reverse=True))   
```

```
strs = ['ccc', 'aaaa', 'd', 'bb']
print(sorted(strs, key=len))   # Sorts by length of the keys 
```

```
# custom Sorting using function . Sorting based on last alphabet
str=['Delhi', 'Kolkata','Hyderabad', 'Bangalore']

def myfun(s):
  return s[-1]

print(sorted(str,key=myfun))
```

# List comprehension

```
# list comprehension is a compact way to write an expression that expands to a whole list
nums = [1, 2, 3, 4]
squares = [ n * n for n in nums ]   ## [1, 4, 9, 16]
print(squares)
```

```
strs = ['hello', 'and', 'goodbye']
shouting = [ s.upper() + '!!!' for s in strs ]
print(shouting)
```

```
# We can add an if test to the right of the for-loop to narrow the result
## Select values <= 2
nums = [2, 8, 1, 6]
small = [ n for n in nums if n <= 2 ]  ## [2, 1]
print(small)
```

```
## Select fruits containing 'a', change to upper case
fruits = ['apple', 'cherry', 'banana', 'lemon']
afruits = [ s.upper() for s in fruits if 'a' in s ]
print(afruits)
## ['APPLE', 'BANANA']
```

# Dictionary

```
# Python's efficient key/value hash table structure is called a "dict". contents of a dict can be written as a series of key:value pairs within braces { }
dict = {}
dict['a'] = 'alpha'
dict['g'] = 'gamma'
dict['o'] = 'omega'

print(dict) ## {'a': 'alpha', 'o': 'omega', 'g': 'gamma'}
```

```
print(dict['a'])    ## Simple lookup, returns 'alpha'
dict['a'] = 6       ## Put new key/value into dict
'a' in dict         ## True
print(dict)
```

```
## By default, iterating over a dict iterates over its keys.
## Note that the keys are in a random order.
for key in dict:
  print(key)
```

```
## Exactly the same as above
for key in dict.keys(): 
  print(key)
## Get the .keys() list:
print(dict.keys()) 
```

```
## Likewise, there's a .values() list of values
print(dict.values())
print('\n')

## Common case -- loop over the keys in sorted order,accessing each key/value
for key in sorted(dict.keys()):
    print(key, dict[key])
print('\n')

## .items() is the dict expressed as (key, value) tuples
print(dict.items())  ##  [('a', 'alpha'), ('o', 'omega'), ('g', 'gamma')]
print('\n')

## This loop syntax accesses the whole dict by looping over the .items() tuple list, accessing one (key, value) pair on each iteration.
for k, v in dict.items(): print(k, '>', v)
```
