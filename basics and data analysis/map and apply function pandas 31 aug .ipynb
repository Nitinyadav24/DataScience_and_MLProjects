```
# apply function 
# to run on a dateset of whole or subset 
import numpy as np
import pandas as pd

df=pd.DataFrame(np.random.normal(size=(4,4)),columns=['A','B','C','D'])
```

```
df
```

```
df.apply(lambda x : 1+ np.abs(x)) # to apply to whole df
df.A.apply(np.abs) # to apply to 1 column
```

```
# apply requires vectorized function 
# for ex for non vectorized 
def double(x):
  if x > 0:
    return 2*x
  return x

```

```
df.apply(double)
```

```
# to get vectorizd 
def double(x):
  x=x.copy()
  x[x>0] *=2
  return x

```

```
a=np.array((1,2,3,4),dtype=int)
```

```
double(a)
```

```
df.apply(double,raw=True) # if raw is true we pas objct as array inseat of series as it enables fast execution 

```
