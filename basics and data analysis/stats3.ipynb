### Poisson Distribution 

```
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

suppose in bank 6 customer arrive every minutes ..what is prob of exactky 4 customer arriving.what is prob of getting more than 4 customers arriving 

```
rate=6
n=np.arange(0,10)
```

```
poisson=stats.poisson.pmf(n,rate)
poisson
```

```
poisson[4]
```

To get the prob more than 4 we subtract from 1 prob of 0,1,2,3 

```
1-(poisson[0]+poisson[1]+poisson[2]+poisson[3])
```

```
plt.plot(n, poisson, 'o-')
plt.show()
```

Binomial talks about defectives , poisson talks about defects within defectives 

when we need to find out defectives go for binomioal  
when need to find out defects per item go for poisson 

Que - Laptop has 12 defects on average 
1.   Calculate if laptop exactly has 5 defects 
2.   Calculate if laptop has less than  5 



```
n=np.arange(0,20)
rate=12
```

```
poisson=stats.poisson.pmf(n,rate)
poisson
```

```
poisson[5]
```

```
poisson[0]+poisson[1]+poisson[2]+poisson[3]+poisson[4]
```

```
plt.plot(n, poisson, 'o-')
plt.show()
```

### Binomial Distribution 

suppose bank wants to determine the prob of customers paying on time . 60% is prob that customer pay on time . If random sample of 7 custom,ers is taken , find the max prob of customer paying on time

```
n=7
```

```
k=np.arange(0,8)
```

```
p=0.6
```

```
binom=stats.binom.pmf(k,n,p)
binom
```

```
plt.plot(k,binom, 'o-')
plt.title('Binomial : n=%i, p=%.2f'%(n,p),fontsize=15)
plt.xlabel('success')
plt.ylabel('prob of sucess')
plt.show()
```

To get cumulative 

```
cum=stats.binom.cdf(k, n,p)
cum
```
