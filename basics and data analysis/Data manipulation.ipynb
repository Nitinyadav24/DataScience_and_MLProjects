### Lets divide the Alumni Salary equally into 3 buckets (Low Salary - L; Medium Salary - M; High Salary - H ) and check which colleges falls in 'H' bucket

```
df1['SG']=pd.cut(df1['Alumni Salary'],3,labels=('L','M','H'))
df1.loc[df1['SG']=='H',:]
# df1['SG'].value_counts()
```
