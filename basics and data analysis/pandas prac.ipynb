## Pre and Post transformation comparison

### Use a Boxplot to compare the runs Scored before and after transformation

```
df['Post Transform']=df['Year']>2012
```

```
df.head()
```

```
sns.boxplot(df['Post Transform'], df['RunsScored'])
```

```
df.head()
```

### Use a Boxplot to compare the Balls Faced before and after transformation

```
sns.boxplot(df['BallsFaced'], df['Post Transform'], orient='h');
```

### Use a Boxplot to compare the runs Scored before and after transformation with Hue as innings(innings wise)

```
sns.boxplot(df['BallsFaced'], df['Post Transform'], hue=df['Inning'],orient='h');
```

```
sns.countplot(df['Dismissal'], hue=df['Post Transform']);
plt.xticks(rotation=90)
```

### Use a bar plot to compare the dismissal patterns before and after transformation

```
ct=pd.crosstab(df['Post Transform'], df['Dismissal'], values=df['RunsScored'], aggfunc='mean')
```

```
ct.plot.bar(figsize=(10,6))
plt.ylabel('Avg Runs Scored')
plt.show()
```

### Use Pointplot to Visualize Yearly trend in run scored by Virat Kohli

```
sns.pointplot(df['Year'],df['RunsScored'], ci=None);
```

### Can we say there is a linear relationship between RunsScored & BallsFaced over different batting positions?

```
plt.figure(figsize=(10,6))
sns.lmplot(x='RunsScored',y='BallsFaced', data=df, hue='BattingPosition');
plt.show()
```

### How the batting position has changed with different oppositions yearly

```
plt.figure(figsize=(10,6))
sns.catplot(x='BattingPosition', 
            y='RunsScored', 
            col='Opposition', 
            data=df, 
            kind='bar',col_wrap=3,ci=None);
```

```
g=sns.FacetGrid(df, col='Opposition', height=5,col_wrap=3);
g=g.map(plt.plot,'Year','BattingPosition', marker='*')
plt.show()
```
