### Functions

```
# Create a function that takes 2 numbers and return the greater one
def greater(a,b):
  if(a>b):
    number=a
  else:
    number=b
  return ('the greater of the number is %d'%number)
```

```
greater(40,50)
```

```
# Create a function to define the odd between 2 numbers
def odd(a):
  if(a%2==0):
    print('the number {} is not odd'.format(a))
  else:
    print(f'the number {a} is odd')
    print('you want to give any other no apart from %d'%a)
```

```
odd(9)
```

```
# Define a function to take the 2 numbers and return the minimum one
def minimum(a,b):
  if(a>b):
    return (f'number {b} is less than {a}')
  elif(a==b):
    print('The no {} and {} are equal'.format(a,b))
  else:
    print('the minimum of the number is %d'%a)
```

```
minimum(5,10)
```

```
# Define a function to find the employee with max working hours using  Tuple unpacking

Hours=[('Billy',400),('George',900),('Tim',1100),('Bob',800)]

def emp_month(Hours):
  max_hour=0
  Emp_name=''
  for employee, work in Hours:
    if work > max_hour:
      max_hour = work
      Emp_name = employee
    else:
      pass
  return (Emp_name,max_hour)
```

```
emp_month(Hours)
```
