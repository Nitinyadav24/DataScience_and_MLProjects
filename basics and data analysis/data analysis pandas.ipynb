### Matrix Multiplication

Matrix Multiplication
The Numpy matmul() function is used to return the matrix product of 2 arrays. Here is how it works

1) 2-D arrays, it returns normal product

2) Dimensions > 2, the product is treated as a stack of matrix

3) 1-D array is first promoted to a matrix, and then the product is calculated

```
print(a * c)
print(np.dot(a,b))
print(a @ b)    #  @ is called mat multi plication 
np.matmul(a,b)  # mat mul and dot are same for 2d arrays but for 3d they are different 
```

```
A =pd.DataFrame(np.random.randn(4,4),columns=(1,2,3,4),index=(1,2,3,4))
B=pd.DataFrame(np.random.rand(4,4),columns=(2,3,4,5),index=(2,3,4,5))
print(A), print(B)
```

```
pd.concat((A,B),axis=0)

pd.concat((A,B),axis=1)

pd.merge(A,B,how='outer',on=2)   # This will outer join union for column 2 , common columns will be name as _x anmd _y 

pd.merge(A,B,how='inner',on=2)    # This will inner join  , common columns will be name as _x anmd _y

pd.merge(A,B,how='left',on=2)     # This will do left join , keep left values , put Nan for right , common columns will be name as _x anmd _y
 
pd.merge(A,B,how='right',on=2)     # This will right join , keep right values put Nan for left , common columns will be name as _x anmd _y
```
