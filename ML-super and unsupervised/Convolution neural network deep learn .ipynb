```
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
```

```
tf.__version__
```

```
# Data preprocessing
# Train Set and Test Set - to avoid overfitting 
# image augmentation - transforming the images so that model dont over learn 
train_datagen=ImageDataGenerator(rescale=1./255,
                                 shear_range=0.2,
                                 zoom_range=0.2,
                                 horizontal_flip=True)
training_set=train_datagen.flow_from_directory(
    'dataset/training_set',
    target_size=(64,64),
    batch_size=32,
    class_mod='binary'
)
```

```
# Preprocessing the test set 
test_datagen=ImageDataGenerator(rescale=1./255,)
test_set=test_datagen.flow_from_directory(
    'dataset/test_set',
    target_size=(64,64),
    batch_size=32,
    class_mod='binary'
)
```

```
# Build the CNN
cnn=tf.keras.models.Sequential()
# convolution layer 
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,activation='relu',input_shape=[64,64,3]))
```

```
# Pooling 
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```
# Adding 2nd convolution layer 
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```
# Flattening - All this conn and pooling to one dimensional vector that becomes the input for fully connected neural network 

cnn.add(tf.keras.layers.Flatten())
```

```
# Full connection - 1 dimension vector 
cnn.add(tf.keras.layers.Dense(units=128,activation='relu'))
```

```
# Outpult layer 
cnn.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```
