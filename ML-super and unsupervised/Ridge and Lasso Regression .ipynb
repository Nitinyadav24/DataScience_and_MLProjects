```
import pandas as pd
```

```
df=pd.read_csv('ridge.csv')
```

```
X=df.iloc[:,:-1]
Y=df.iloc[:,-1]
```

### Ridge 

```
from sklearn.linear_model import Ridge
```

```
ridge_l=[0,1,10,100]

for i, l in enumerate(ridge_l):
  ridge=Ridge(alpha=l)
  ridge.fit(X,Y)
  
  ridge_coeff=ridge.coef_
  ridge_int=ridge.intercept_

  x_plt=[0,1,2,3,4,5]
  Y_plt=ridge.predict(pd.DataFrame(x_plt))


  plt.figure(1)
  plt.subplot(2,2,i+1)
  plt.plot(x_plt,Y_plt)
  plt.ylim(ymin=0, ymax=9)
  plt.xlim(xmin=0, xmax=6)
  plt.title('y =' + str('%.2f' %ridge_coeff)+ '* x + ' + str('%.2f' %ridge_int) + ' for \u03BB or \u03B1 =' + str(l),fontsize=12) 
  plt.tight_layout()
```

### LAsso 

```
X=df.iloc[:,:-1]
Y=df.iloc[:,-1]
```

```
from sklearn.linear_model import Lasso
```

```
lasso_l=[0,0.5,1,2,4]

for i, l in enumerate(lasso_l):
  lasso=Lasso(alpha=l)
  lasso.fit(X,Y)
  
  lasso_coeff=lasso.coef_
  lasso_int=lasso.intercept_

  x_plt=[0,1,2,3,4,5]
  Y_plt=lasso.predict(pd.DataFrame(x_plt))


  plt.figure(2)
  
  plt.plot(x_plt,Y_plt)
  plt.ylim(ymin=0, ymax=9)
  plt.xlim(xmin=0, xmax=6)
  plt.text(x_plt[-1],Y_plt[-1],'y =' + str('%.2f' %lasso_coeff)+ '* x + ' + str('%.2f' %lasso_int) + ' for \u03BB or \u03B1 =' + str(l),fontsize=12) 
  plt.tight_layout()
```
