### Plotting categorical along with continuous 

```
sns.stripplot(df['fuel_type'],df['horsepower'], jitter=True);
```

```
# We ue swarm plot to show the spread of data 
sns.swarmplot(df['fuel_type'],df['horsepower']);
```

```
sns.boxplot(df['number_of_doors'],df['horsepower']);
```

```
sns.boxplot(df['number_of_doors'],df['horsepower'], hue=df['fuel_type']);
```

```
sns.barplot(df['body_style'],df['horsepower'],hue=df['fuel_type']);
```

```
# Bar plot above plots the bar for each category which shows the average value
# if we want to count the number or each body style then we have to use count plot 
# The black line shows the 95% confidence interval with the lower and upper bounds 

sns.countplot(df['body_style'],hue=df['fuel_type']);
```

```
# Making the bar to absolutely point to zero is misleading so we use the point plot 
sns.pointplot(df['body_style'],df['horsepower'],hue=df['fuel_type']);
```

#### Multi panel categorical plots 

```
sns.catplot(x='fuel_type',
            y='horsepower',
            hue='number_of_doors',
            col='drive_wheels',
            data=df,kind='box');
```

```
sns.catplot(x='fuel_type',
            y='horsepower',
            hue='number_of_doors',
            col='drive_wheels',
            data=df,kind='violin');
```

```
sns.catplot(x='fuel_type',
            hue='number_of_doors',
            col='drive_wheels',
            data=df,kind='count');
```

```
sns.catplot(x='fuel_type',
            y='horsepower',
            hue='number_of_doors',
            col='drive_wheels',
            data=df,kind='strip');
```

### Linear model plot 

```
sns.lmplot(y='horsepower',x='engine_size', hue='fuel_type', data =df);
```

```
# line plot 
x=np.arange(0,10)
y=np.linspace(1,10,10)
```

```
plt.plot(x,y, color='r', linewidth=2, linestyle=':')
plt.plot(x,y*2, color='r', linewidth=2, linestyle='-.')
plt.show()
```

```
### Creating Subplots , here (1,2,1) in subplot means 1 row , 2 columns and first plot
plt.subplot(1,2,1)
plt.plot(x,y, color='r', linewidth=2, linestyle=':')
plt.subplot(1,2,2)  # here 1,2,2 means 1 row 2 columns and 2nd plot
plt.plot(x,y*2, color='r', linewidth=2, linestyle='-.')
plt.show()
```

```
plt.subplot(2,1,1)
plt.plot(x,y, color='r', linewidth=2, linestyle=':')
plt.subplot(2,1,2)
plt.plot(x,y*2, color='r', linewidth=2, linestyle='-.')
plt.show()
```

```
## Bar Plot 
dict={'A':40,'B':30, 'C':55,'D':78}
name=dict.keys()
name=list(name)
value=list(dict.values())
```

```
plt.barh(name,value, color=('g','r','b','r'))
plt.grid(True)
plt.show()
```
