```
import numpy as np
import pandas as pd
```

```
np.random.seed(5)
x=np.linspace(1,30,10)
```

```
y1=x/(np.random.rand(10)*2)
y2=x/(np.random.rand(10)*3)
```

```
from matplotlib.lines import Line2D
```

```
line1=Line2D(x,y1)
line2=Line2D(x,y2)
fig,ax=plt.subplots()
ax.add_line(line1)
ax.add_line(line2)

ax.autoscale()
```

```
fig.set_size_inches(15,7)
fig
```

```
l1=ax.lines[0]
l2=ax.lines[1]

l1.set_alpha(0.4)
fig
```

```
# to set the properties of line 
for line in ax.lines:
  line.set_color('r')
  line.set_lw(3)
fig
```

```
l1.set_linestyle('dashed')
fig
```

```
l2.set_marker('*')
l2.set_markersize(20)
l2.set_markerfacecolor('white')
l2.set_markevery(1)
l2.set_sketch_params(1)
fig
```

```
ax.set_title('some title')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
fig
```

```
fig.savefig('artist.png',dpi=350,transparent=False)
```

```
fig1,ax1=plt.subplots()
ax1.plot(x,y1)
ax1.plot(x,y2)
```
