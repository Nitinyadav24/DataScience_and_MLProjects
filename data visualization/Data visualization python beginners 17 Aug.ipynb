```
import matplotlib.pyplot as plt
with open('Goals.txt','r') as f:
  HomeTeamGoal=[int(x) for x in f.readline().strip('\n').strip(' ').split()]
  AwayTeamGoal=[int(x) for x in f.readline().strip('\n').strip(' ').split()]
```

```
x=[]
for i in range(len(HomeTeamGoal)):
  x.append(i)
# Adjusting figure size and labels 
plt.figure(figsize=(10,6))
plt.plot(x,HomeTeamGoal,c='blue',ls='--',label='Home Goal')
# Ading title , xlabel and y label
plt.title('our first plot')
# Adding lawtech unique in labels
ticks =['Game One','Game twenty', 'Game 100']
plt.xlabel(r'Game $_5$')
plt.ylabel('Goals')
plt.plot(x,AwayTeamGoal,c='green', ls=':',label='Away Goal')
plt.legend(loc='upper right')
plt.xlim(0,145)
plt.ylim(0,6)
# Adding custom ticks in the graph
plt.xticks([0,49,99],ticks,rotation=45)
plt.yticks([4],['Four'])
# Adding Text
plt.text(50,4,'Text',fontsize=10,c='Red')
plt.show()
# put different line styles ls='--' for dash , ':' for dotted 


```

```
TotalGoals=HomeTeamGoal+AwayTeamGoal
plt.figure(figsize=(8,8))
# to compare two values put in tuple under value X , also add labels 
plt.hist(x=(HomeTeamGoal,AwayTeamGoal,TotalGoals),bins=range(8),label=['Home Goal','Away Goal','TotalGoal'],align='left')
plt.legend()
plt.xlabel('Goal score')
plt.ylabel('No of Goal')
plt.show()
```

```
plt.figure(figsize=(8,8))
plt.hist2d(x=HomeTeamGoal,y=AwayTeamGoal,
          bins=(range(8),range(7)))
plt.xlabel('Home Goal')
plt.ylabel('Away Goal')
plt.show()
```
