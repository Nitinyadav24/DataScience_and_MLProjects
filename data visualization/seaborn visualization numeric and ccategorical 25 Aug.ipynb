```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
%matplotlib inline
df=pd.read_csv('/content/Automobile (1).csv')
```

```
sns.distplot(df['highway_mpg'])
```

```
sns.distplot(df['city_mpg'],kde=False,rug=True)
```

```
# plotting 2d plots
# in seaborn scatter is called joint plot
sns.jointplot(df['engine_size'],df['horsepower']);
```

```
sns.jointplot(df['engine_size'],df['horsepower'],kind='hex');
```

```
sns.jointplot(df['engine_size'],df['horsepower'],kind='kde')
```

```
# to plot multiple pairwise scatter we use pairplot function
sns.pairplot(df[['normalized_losses','engine_size','horsepower']]);
```

```
## Graphs for discrete variables 
sns.stripplot(df['fuel_type'],df['horsepower'],jitter=True)
```

```
# To avoid data points overlapping , we use swarm plot
sns.swarmplot(df['fuel_type'],df['horsepower'])
```

```
# Box plots
sns.boxplot(df['number_of_doors'],df['horsepower'],hue=df['fuel_type']);
```

```
Df2=df[df['number_of_doors']=='two']
```

```
Df2['horsepower'].max()
```

```
# Bar plot
sns.barplot(df['body_style'],df['horsepower'],hue=df['fuel_type'])
# Black line shows confidence interval
```

```
# Want to know the count of category rather than calculating the mean 
sns.countplot(df['body_style'],hue=df['fuel_type'])
```

```
# pointplot
sns.pointplot(df['body_style'],df['horsepower'],hue=df['number_of_doors'])
```

```
# Multipnanel categorical plot using Catplot
sns.catplot(x='fuel_type',
y='horsepower',
hue='number_of_doors',
col='drive_wheels',
data=df,
kind='violin')
# In this kind we can have {point,bar,count,box,violin,strip}
```
